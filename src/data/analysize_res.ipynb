```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.genfromtxt('perf_python.csv', delimiter=' ', skip_header=1)
print (data)
```

```python
x = [10**i for i in range(7)]
print (x)
projection = data[:, 0]
grow = data[:, 1]
casting = data[:, 2]
```

```python
fig, ax = plt.subplots()
ax.loglog(x, projection, label='projection')
ax.loglog(x, grow, label='grow')
ax.loglog(x, casting, label='casting')
ax.set_xlabel('Number of keys')
ax.set_ylabel('Time (ms)')
ax.legend()

# show the y labels in float format
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: '%1.1f' % x))

plt.grid(True)
plt.savefig('perf_python.png')
plt.show()
```
